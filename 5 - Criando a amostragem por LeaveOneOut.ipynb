```python
#importando pacotes úteis
import pandas as pd
import numpy as np
from sklearn.model_selection import LeaveOneOut
```

```python
dt_agreg = pd.read_csv('dt_agreg.csv')
dt_agreg
```

```python
loo = LeaveOneOut()
dt_data = dt_agreg.copy()
dt_data.drop(columns=['Category'])
dt_target = dt_agreg['Category']
loo.get_n_splits(dt_data)
```

```python
for train_index, test_index in loo.split(dt_agreg):
    print("TRAIN:", train_index)
    print("TEST:", test_index)
    dt_data_train = dt_data.iloc[train_index]
    dt_target_train = dt_target[train_index]
    dt_data_test = dt_data.iloc[test_index]
    dt_target_test = dt_target[test_index]
    print("DATASET_TRAIN: ", dt_data_train)
    print("TARGET_TRAIN: ", dt_target_train)
    print("DATASET_TEST: ", dt_data_test)
    print("TARGET_TEST: ", dt_target_test)
```
