```python
#importando pacotes úteis
import pandas as pd
import numpy as np
import random
```

```python
dt_2apli = pd.read_csv('2apli.csv')
dt_3apli = pd.read_csv('3apli.csv')
```

```python
n_cat = 5
dict_cat = {
    'Muito baixo':0,
    'Baixo':1,
    'Medio':2,
    'Alto':3,
    'Muito alto': 4
}
dict_cat_inv = {
    0:'Muito baixo',
    1:'Baixo',
    2:'Medio',
    3:'Alto',
    4:'Muito alto'
}
```

```python
#25% de ruido com 25% de incerteza
dt_2apli_ruido = dt_2apli.copy()
for index in dt_2apli_ruido.sample(frac=0.25, random_state=0).index:
    for column in dt_2apli_ruido.loc[:, dt_2apli_ruido.columns != 'Nivel de interferencia'].sample(frac=0.25, axis=1, random_state=index).columns:
        dt_2apli_ruido.iloc[index][column] = dict_cat_inv[(dict_cat[dt_2apli_ruido.iloc[index][column]]+random.choice([-1,1]))%n_cat]
dt_3apli_ruido = dt_3apli.copy()
for index in dt_3apli_ruido.sample(frac=0.25, random_state=0).index:
    for column in dt_3apli_ruido.loc[:, dt_3apli_ruido.columns != 'Nivel de interferencia'].sample(frac=0.25, axis=1, random_state=index).columns:
        dt_3apli_ruido.iloc[index][column] = dict_cat_inv[(dict_cat[dt_3apli_ruido.iloc[index][column]]+random.choice([-1,1]))%n_cat]
dt_2apli_ruido.to_csv('2apli_25ruido_25incerteza.csv', index=False)
dt_3apli_ruido.to_csv('3apli_25ruido_25incerteza.csv', index=False)
```
