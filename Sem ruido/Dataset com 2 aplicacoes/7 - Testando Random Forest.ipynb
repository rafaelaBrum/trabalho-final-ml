```python
#importando pacotes úteis
import pandas as pd
import numpy as np
from sklearn.model_selection import LeaveOneOut
from sklearn.ensemble import RandomForestClassifier
```

```python
dt_2apli_nominal = pd.read_csv('2apli.csv');
dt_2apli_nominal
```

```python
dt_2apli_nominal['Nivel de interferencia'].value_counts()
```

```python
erro_majoritario = dt_2apli_nominal['Nivel de interferencia'].value_counts().max()/dt_2apli_nominal['Nivel de interferencia'].count()
print("erro_majoritario = ", erro_majoritario) 
```

```python
dict_cat = {
    'Muito baixo':0,
    'Baixo':1,
    'Medio':2,
    'Alto':3,
    'Muito alto': 4,
    'Muito baixa':0,
    'Baixa':1,
    'Media':2,
    'Alta':3,
    'Muito alta': 4
}
```

```python
#O Random Forest só funciona com valores numéricos
dt_2apli = pd.DataFrame(columns=['App1 - SLLC', 'App1 - DRAM', 'App1 - NET', 'App2 - SLLC', 'App2 - DRAM', 'App2 - NET'])
dt_2apli['App1 - SLLC'] = dt_2apli_nominal['App1 - SLLC'].map(dict_cat)
dt_2apli['App1 - DRAM'] = dt_2apli_nominal['App1 - DRAM'].map(dict_cat)
dt_2apli['App1 - NET'] = dt_2apli_nominal['App1 - NET'].map(dict_cat)
dt_2apli['App2 - SLLC'] = dt_2apli_nominal['App2 - SLLC'].map(dict_cat)
dt_2apli['App2 - DRAM'] = dt_2apli_nominal['App2 - DRAM'].map(dict_cat)
dt_2apli['App2 - NET'] = dt_2apli_nominal['App2 - NET'].map(dict_cat)
dt_category = dt_2apli_nominal['Nivel de interferencia'].map(dict_cat)
dt_target = np.array(dt_category)
dt_data = np.array(dt_2apli)
print('dt_target = ', dt_target)
print('dt_data = ', dt_data)
```

```python
loo = LeaveOneOut()
loo.get_n_splits(dt_data)
count_rights = 0
count_wrongs = 0
for train_index, test_index in loo.split(dt_data):
    print("Test_index = ", test_index)
    dt_data_train = dt_data[train_index]
    dt_target_train = dt_target[train_index]
    dt_data_test = dt_data[test_index]
    dt_target_test = dt_target[test_index]
    clf = RandomForestClassifier(n_estimators=100, random_state=0, bootstrap=False) #bootstrap default é True
    clf.fit(dt_data_train, dt_target_train)
        
    dt_z_test = clf.predict(dt_data_test)
    
    if(dt_z_test == dt_target_test):
        #print("acertou")
        count_rights = count_rights + 1
    else:
        #print("errou")
        count_wrongs = count_wrongs + 1
```

```python
print("Acertos = ", count_rights)
print("Erros = ", count_wrongs)
```

```python
n_elements = count_rights + count_wrongs
precision = count_rights/n_elements
print("n_elements = ", n_elements)
print("precision = ", precision)
#precision = np.true_divide(count_rights,215)
#precision 0,8008849557522124
```
