```python
qtde_duplicados = 0
qtde_ruidos = 0
for cat_sllc_app1 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
    for cat_dram_app1 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
        for cat_net_app1 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
            for cat_sllc_app2 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
                for cat_dram_app2 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
                    for cat_net_app2 in ['Muito baixo', 'Baixo', 'Medio', 'Muito alto']:
                        #print("sllc_app1=", cat_sllc_app1, "-dram_app1=", cat_dram_app1, "-net_app1=", cat_net_app1, "-sllc_app2=", cat_sllc_app2, "-dram_app2=", cat_dram_app2, "-net_app2=", cat_net_app2)
                        dt_duplicados = dt_2apli.loc[(dt_2apli['App1 - SLLC'] == cat_sllc_app1) & 
                                                     (dt_2apli['App1 - DRAM'] == cat_dram_app1) & 
                                                     (dt_2apli['App1 - NET'] == cat_net_app1) & 
                                                     (dt_2apli['App2 - SLLC'] == cat_sllc_app2) & 
                                                     (dt_2apli['App2 - DRAM'] == cat_dram_app2) & 
                                                     (dt_2apli['App2 - NET'] == cat_net_app2)]
                        if(not dt_duplicados.empty) and (dt_duplicados['Nivel de interferencia'].count() > 1):
                            #print("sllc_app1=", cat_sllc_app1, "-dram_app1=", cat_dram_app1, "-net_app1=", cat_net_app1, "-sllc_app2=", cat_sllc_app2, "-dram_app2=", cat_dram_app2, "-net_app2=", cat_net_app2, "-qtde_linhas=", dt_duplicados.count())
                            print("Temos mais de um com os atributos de entrada iguais!!", "-qtde_linhas=", dt_duplicados['Nivel de interferencia'].count())
                            print(dt_duplicados['Nivel de interferencia'].value_counts())
                            qtde_duplicados = qtde_duplicados + (dt_duplicados['Nivel de interferencia'].value_counts())[0] -1
                            qtde_ruidos = qtde_ruidos + (dt_duplicados['Nivel de interferencia'].count() - (dt_duplicados['Nivel de interferencia'].value_counts())[0])
print("qtde_duplicados = ", qtde_duplicados)
print("qtde_ruidos = ", qtde_ruidos)
```
